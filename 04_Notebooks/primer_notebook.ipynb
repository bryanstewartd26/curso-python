```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos simulados de ventas de zapatos
data = {
    'Producto': ['Tenis A', 'Tenis B', 'Tenis C', 'Tenis D', 'Tenis E'],
    'Ventas_Mes1': [120, 80, 150, 60, 200],
    'Ventas_Mes2': [130, 95, 160, 70, 210]
}

# Convertimos los datos a tabla
df = pd.DataFrame(data)
df
```

```python
# Calcular el total de ventas por producto
df['Total'] = df['Ventas_Mes1'] + df['Ventas_Mes2']

# Mostrar el producto más vendido
producto_top = df.loc[df['Total'].idxmax()]
print("🏆 El producto más vendido es:", producto_top['Producto'])
```

```python
# Crear un gráfico de barras
plt.bar(df['Producto'], df['Total'], color='orange')
plt.title('Ventas Totales por Producto 👟')
plt.xlabel('Producto')
plt.ylabel('Unidades Vendidas')
plt.show()
```
